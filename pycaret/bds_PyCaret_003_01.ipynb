 # **<font color="DeepPink">PyCaret - Clustering</font>**

<p align="justify">
En <code>PyCaret</code>, el módulo de Clustering (Clustering Module) es una parte esencial de la biblioteca diseñada para realizar tareas de agrupamiento automatizado. El agrupamiento, también conocido como clustering en inglés, es una técnica de aprendizaje automático no supervisado que tiene como objetivo dividir un conjunto de datos en grupos o clústeres, donde los elementos dentro de cada clúster son similares entre sí y diferentes de los elementos en otros clústeres.
<br><br>
El módulo de Clustering de <code>PyCaret</code> ofrece una serie de funciones y herramientas que facilitan el proceso de agrupamiento de datos de manera automatizada y eficiente.
<br><br>
Algunas de las características clave del módulo de Clustering de <code>PyCaret</code> son:
<br>
<br>
<ol align="justify">
<li><b>Selección de algoritmo</b>: <code>PyCaret</code> admite una variedad de algoritmos de agrupamiento populares, como K-Means, DBSCAN, Mean Shift, y más. Los usuarios pueden elegir entre diferentes algoritmos para encontrar el que mejor se ajuste a sus datos y objetivos.</li>
<br>
<li><b>Optimización de hiperparámetros</b>: Al igual que en los módulos de Clasificación y Regresión, <code>PyCaret</code> realiza una búsqueda inteligente en el espacio de hiperparámetros para encontrar la configuración óptima para clustering.</li>
<br>
<li><b>Visualización de clústeres</b>: <code>PyCaret</code> proporciona herramientas para visualizar los resultados del agrupamiento, lo que ayuda a comprender cómo se han formado los clústeres y cómo se distribuyen los datos dentro de ellos.</li>
<br>
<li><b>Análisis y selección de características</b>: <code>PyCaret</code> permite a los usuarios realizar análisis y selección de características antes de aplicar el algoritmo de agrupamiento, lo que ayuda a mejorar la calidad y la interpretación de los resultados.</li>
<br>
<li><b>Evaluación del agrupamiento</b>: <code>PyCaret</code> proporciona métricas de evaluación del agrupamiento, como  Silueta (Silhouette) y la Inercia (Inertia), que ayudan a evaluar la calidad de los clústeres generados.</li>
<br>
<li><b>Predicción en nuevos datos</b>: Una vez que se ha realizado el agrupamiento en el conjunto de entrenamiento, PyCaret permite aplicar el modelo de agrupamiento a nuevos datos para asignarlos a los clústeres existentes.</li>
</ol>
<br>
<p align="justify">
El módulo de Clustering de <code>PyCaret</code> es una herramienta valiosa para estudiantes y profesionales de Ciencia de Datos que deseen explorar y comprender patrones y estructuras ocultas en sus datos sin la necesidad de supervisión. Al automatizar gran parte del proceso de agrupamiento, <code>PyCaret</code> permite a los usuarios centrarse en el análisis e interpretación de los resultados, lo que facilita la toma de decisiones informadas y la obtención de información valiosa de los datos no etiquetados.

 ### **<font color="DeepPink">Silhouette</font>**

<p align="justify">
El coeficiente de Silhouette, comúnmente conocido como "Silhouette", es una métrica utilizada para evaluar la calidad de un  cluster de datos. Se utiliza en técnicas de aprendizaje no supervisado, como el clustering, para medir qué tan bien los datos están agrupados y separados unos de otros.
<br><br>
El coeficiente de Silhouette se calcula para cada punto de datos y proporciona una puntuación que va desde $-1$ hasta $1$. Los valores más altos de Silhouette indican que un punto de datos está bien agrupado, mientras que valores cercanos a $-1$ indican que el punto de datos podría haber sido asignado a un grupo incorrecto.
<br><br>
El cálculo del coeficiente de Silhouette implica tres componentes principales:
<br><br>
<ol align="justify">
<li><b>Distancia intra-cluster (a)</b>: Es la media de la distancia entre un punto y todos los demás puntos en el mismo cluster al que pertenece. Representa qué tan compacto está el punto dentro de su cluster.</li>
<br>
<li><b>Distancia media al cluster más cercano (b)</b>: Es la media de la distancia entre un punto y todos los puntos en el cluster más cercano diferente al que pertenece. Representa qué tan separado está el punto de otros clusters.</li>
<br>
<li><b>Coeficiente de Silhouette (s)</b>: Se calcula para cada punto usando la siguiente fórmula:</li>
<br><br>
 $$s = \frac{b - a}{\max(a, b)}$$
<br>
Una vez calculados los coeficientes de Silhouette para todos los puntos, se puede tomar el promedio de todos los coeficientes para obtener la puntuación general del agrupamiento.</ol>
<br>
<p align="justify">
Interpretación del coeficiente de Silhouette:
<br>
<ul align="justify">
<li> Si el coeficiente de Silhouette es cercano a $1$, significa que el punto está bien agrupado y separado de otros clusters, lo que indica un buen agrupamiento.</li>
<br>
<li> Si el coeficiente de Silhouette es cercano a $0$, indica que el punto está cerca del límite entre dos clusters y podría haber sido asignado a diferentes clusters, lo que sugiere una ambigüedad en la agrupación.</li>
<br>
<li> Si el coeficiente de Silhouette es cercano a $-1$, indica que el punto está más cerca de un cluster diferente al que pertenece, lo que sugiere un mal agrupamiento.</li></ul><br>
<p align="justify">
En resumen, el coeficiente de Silhouette es una medida útil para evaluar la calidad del agrupamiento de datos y seleccionar el número óptimo de clusters en técnicas de clustering.
<br><br>
Un valor alto de Silhouette indica que el agrupamiento es más sólido y confiable, mientras que un valor bajo sugiere que puede haber problemas con la separación de los datos en grupos coherentes.

 ### **<font color="DeepPink">Calinski-Harabasz</font>**

<p align="justify">
El índice de Calinski-Harabasz, también conocido como índice de varianza entre grupos (Variance Ratio Criterion), es una métrica utilizada para evaluar la calidad de un agrupamiento (cluster) de datos en técnicas de aprendizaje no supervisado, como el clustering.
<br><br>
El índice de Calinski-Harabasz se basa en la idea de que un buen agrupamiento debería tener una alta dispersión entre los grupos (clusters) y una baja dispersión dentro de los grupos. Es decir, los puntos dentro de cada grupo deberían estar bien agrupados y separados claramente de los puntos en otros grupos.
<br><br>
El cálculo del índice de Calinski-Harabasz implica tres componentes principales:
<br><br>
<ol align="justify">
<li><b>Dispersiones entre grupos (Between-Cluster Dispersion, B)</b>: Es una medida de qué tan separados están los grupos entre sí. Se calcula sumando la dispersión entre el centroide global de los datos y los centroides de cada grupo, y luego multiplicándolo por el número de puntos en el conjunto de datos.</li>
<br>
<li><b>Dispersiones dentro de grupos (Within-Cluster Dispersion, W)</b>: Es una medida de qué tan compactos están los puntos dentro de cada grupo. Se calcula sumando la dispersión dentro de cada grupo, que es la suma de las distancias entre cada punto y el centroide del grupo al que pertenece.</li>
<br>
<li><b>Índice de Calinski-Harabasz (CH)</b>: Se calcula usando la siguiente fórmula:</li>
<br>
 $$CH = \frac{B}{W} \times \frac{N - K}{K - 1}$$
<br>
Donde:
<br><ul align="justify">
<li> \( N \) es el número total de puntos en el conjunto de datos.</li>
<li> \( K \) es el número de clusters.</li>
</ul></ol>
<p align="justify">
El índice de Calinski-Harabasz busca maximizar la relación entre la dispersión entre grupos y la dispersión dentro de los grupos. Cuanto mayor sea el valor de CH, mejor será la calidad del agrupamiento. Un valor alto de CH indica que los grupos están bien separados y compactos, lo que sugiere que el agrupamiento es sólido y coherente.
<br>
Es importante tener en cuenta que el índice de Calinski-Harabasz puede ser sensible al tamaño y forma de los clusters, por lo que debe utilizarse en conjunto con otras métricas y técnicas para evaluar y validar el agrupamiento en un problema específico.
<br><br>
En general, el índice de Calinski-Harabasz es una herramienta útil para seleccionar el número óptimo de clusters y evaluar la calidad del agrupamiento en técnicas de clustering.

 ### **<font color="DeepPink">Davies-Bouldin</font>**

<p align="justify">
El índice Davies-Bouldin es otra métrica utilizada para evaluar la calidad de un  cluster de datos en técnicas de aprendizaje no supervisado, como el clustering. Al igual que el índice de Calinski-Harabasz, el índice Davies-Bouldin busca medir qué tan bien están agrupados los datos y cuán separados están los grupos entre sí.
<br><br>
El cálculo del índice Davies-Bouldin se basa en la idea de que un buen agrupamiento debería tener grupos bien definidos y claramente separados entre sí.
<br><br>
Para calcular el índice, se realiza lo siguiente para cada par de clusters $i$ y $j$:
<br><br>
<ol align="justify">
<li>Se calcula la dispersión dentro del cluster \( i \) (denotada como \( d_i \)) como la distancia promedio entre cada punto en el cluster \( i \) y el centroide del mismo.</li>
<br>
<li>Se calcula la dispersión dentro del cluster \( j \) (denotada como \( d_j \)) de manera similar.</li>
<br>
<li>Se calcula la distancia entre los centroides de los clusters \( i \) y \( j \) (denotada como \( d_{ij} \)).</li>
<br>
<li>Se calcula el índice Davies-Bouldin para el par de clusters \( i \) y \( j \) como:
<br><br>
\[ R_{ij} = \frac{d_i + d_j}{d_{ij}} \]
<br></li>
<li>Se repite este proceso para todas las combinaciones de clusters y se obtiene el valor promedio de \( R_{ij} \) para todos los pares de clusters. El valor promedio se toma como el índice Davies-Bouldin para el agrupamiento completo.</li></ol>
<br><p align="justify">
Un índice Davies-Bouldin más bajo indica un mejor agrupamiento, ya que significa que los grupos están más claramente separados y más compactos dentro de sí mismos. Por lo tanto, al comparar diferentes agrupamientos, aquellos que tengan un índice Davies-Bouldin más bajo se consideran más deseables.
<br><br>
Al igual que con otras métricas de evaluación de clustering, el índice Davies-Bouldin debe utilizarse junto con otras métricas y técnicas para obtener una comprensión completa de la calidad del agrupamiento y seleccionar el número óptimo de clusters en función de la estructura de los datos.
