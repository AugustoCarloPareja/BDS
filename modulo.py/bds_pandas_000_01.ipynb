````python
import pandas as pd
````

## **<font color="DeepPink">Configurando las opciones deseadas</font>**

````python
pd.options.display.precision = 2
pd.options.display.float_format = "{:.2f}".format
````

## **<font color="DeepPink">Obtenci√≥n de datos</font>**



<p align="justify"><b>
<font color="DeepPink">
Buscamos la p√°gina web en donde se encuentran los datos que queremos adquirir. Luego, al objeto <code>data</code> le asignamos la direcci√≥n URL de la p√°gina web de la Administraci√≥n Tributaria Mendoza.
</font>
</b></p>


https://www.atm.mendoza.gov.ar/portalatm/zoneBottom/datosInteres/recaudacion/recaudacion_por_impuesto_anual.jsp

````python
data = "https://www.atm.mendoza.gov.ar/portalatm/zoneBottom/datosInteres/recaudacion/recaudacion_por_impuesto_anual.jsp"
````

## **<font color="DeepPink">Creando un dataframe</font>**


### **<font color="DeepPink">Leyendo los datos</font>**

<p align="justify"><b>
<font color="DeepPink">
Utilizamos el m√©todo de Pandas <code>read_html()</code> para leer los datos de la p√°gina web, y utilizamos los par√°metros <code>index_col=0</code> y <code>header=1</code> para hacer un poco legigle los datos...
</font>
</b></p>


````python
lista = pd.read_html(data, flavor="html5lib", index_col=0, header=1)
````

<p align="justify">
<font color="DeepPink">

**Se procede a verificar el objeto creado con el m√©todo ```type()```...**

</font>
</p>

````python
type(lista)
````

### **<font color="DeepPink">Visualizando los datos</font>**

<p align="justify">
<font color="DeepPink">

**Verificamos la cantidad de elementos que tiene la lista con el m√©todo ```len()``` y luego, llamamos al objeto ```df``` para ver los datos...**

</font>
</p>

````python
len(lista)
````

<p align="justify"><b>
<font color="DeepPink">
El objeto tiene $23$ elementos, que corresponden a las diferentes tablas con las recaudaciones de los impuestos por a√±o fiscal, desde el a√±o $2000$ hasta el a√±o $2022$...
</font>
</b></p>

### **<font color="DeepPink">Dando el formato de dataframe</font>**

<p align="justify"><b>
<font color="DeepPink">
Primero vamos a visualizar el √∫ltimo elemento de la lista, ya sabiendo que la lista posee $23$ elementos, entonces llamamos al √∫ltimo...

</font>
</p>

````python
lista[22]
````

<p align="justify"><b>
<font color="DeepPink">
El elemento tiene aspecto de dataframe, por lo cual podemos inferir que la lista, es una lista de dataframes, y cada uno de sus elementos es un dataframe correspondiente a cada uno de los a√±os de recaudaci√≥n, desde el a√±o $2000$ hasta el a√±o actual. Por lo tanto, el √∫ltimo elemento de la lista, es el a√±o actual.
</font>
</p>

<p align="justify">
<font color="DeepPink">

> **Lo primero que vamos a hacer es eliminar las columnas con datos faltantes...**
</font>
</p>

````python
df = lista[22].dropna(axis=1, how="all")
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Ahora vamos a renombrar las columnas del dataframe, con los datos de la primera fila, que son los impuestos provinciales...**
</font>
</p>

````python
df.columns = df.iloc[0,]
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Vemos que las columnas tienen por nombre ```nan```, vamos a eliminar ese nombre y no dejar ning√∫n nombre a las columnas...**
</font>
</p>

````python
df.columns.name = None
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Ahora al indice del dataframe, en donde estan los meses del ejercicio fiscal, le vamos a poner un nombre, para luego, ingresarlo como columna al dataframe...**
</font>
</p>

````python
df.index.name = "Datos"
````

````python
df.reset_index(inplace=True)
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Ahora vamos a eliminar las filas que tengan todos sus datos faltantes, para ello usamos el m√©todo de Pandas ```dropna()``` con los par√°metros ```axis=0``` para indicar que queremos eliminar las filas, y el parametro ```how=all```, es decir eliminar las filas en donde todos sus datos son faltantes...**
</font>
</p>

````python
df = df.dropna(axis=0, how="all")
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Lo que vamos a hacer ahora, es ordenar el indice, usando el m√©todo ```reset_index()```, eliminando el √≠ndice, sin ingresarlo al dataframe, y con posterioridad a ello, vamos a eliminar la primera fila y la √∫ltima fila del dataframe, pero con el uso de filtros...**
</font>
</p>

````python
df.reset_index(drop=True, inplace=True)
````

````python
df
````

````python
df = df[(df.Automotor != "Automotor") & (df.Datos != "Totales")]
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Ahora procedemos a borrar la columna ```Datos```...**
</font>
</p>

````python
df = df.drop(columns="Datos")
````

````python
df
````

````python
df.info()
````

<p align="justify">
<font color="DeepPink">

> **Ahora vamos a convertir todos los datos del dataframe, en num√©ricos decimales...**
</font>
</p>

````python
df = df.astype("float")
````

````python
df.info()
````

<p align="justify">
<font color="DeepPink">

> **Ahora al indice num√©rico del dataframe, coincidente con los meses del ejercicio fiscal, le vamos a poner un nombre, para luego, ingresarlo como columna al dataframe...**
</font>
</p>

````python
df.index.name = "Mes"
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Ahora ingresamos el indice del dataframe, como columna...**
</font>
</p>

````python
df.reset_index(inplace=True)
````

````python
df
````

<p align="justify">
<font color="DeepPink">

> **Y vamos adicionar una columna denominada ```A√±o``` que tenga como valor, el a√±o fiscal correspondiente a este dataframe, que en este caso es el a√±o $2022$...**
</font>
</p>

````python
df["A√±o"] = int(2022)
````

````python
df
````

````python
df.info()
````

<p align="justify">
<font color="DeepPink">

> **Y ordenamos las columnas...**
</font>
</p>

````python
df = df[["Mes","A√±o","Ingresos Varios","Ingresos Brutos","Inmobiliario","Automotor","Sellos"]]
````

````python
df
````

<p align="justify"><b>
<font color="DeepPink">
‚úÖ Ya tenemos nuestro dataframe correspondiente al ejercicio fiscal $2022$ con los datos num√©ricos.
</font>
</p>
